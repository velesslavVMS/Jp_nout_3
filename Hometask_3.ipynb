```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

3.1 Найдите среднюю стоимость домов в зависимости от количества спален. Отсортируйте от меньшей стоимости к большей

```python
df['bedrooms'].value_counts()
```

```python
bedgroup = df.groupby('bedrooms')
groupPrice= bedgroup.agg({'price':'mean'})
round(groupPrice.sort_values('price'))
```

3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома  condition

```python
condGroup = df.groupby('condition')
condPrice = condGroup.agg({'price': ['min','mean','max']})
round(condPrice)
```

3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df['one'] = 1
df.head()
```

```python
df.pivot_table(index = 'view',columns = 'waterfront', values = 'one', aggfunc = 'sum', margins = True)
```

3.4 Каких домов в зависимости от этажности и количества спален больше?
Больше всего одноэтажных домов с тремя спальнями.

```python
pd.crosstab(index = df['bedrooms'], columns = df['floors'],values = df['one'], aggfunc= 'sum')
```

3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index = df['grade'], columns = df['condition'], values = df['price'],aggfunc='median' )
```
